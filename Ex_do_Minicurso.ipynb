Vamos inicia importando duas Bibliotecas: Pandas e Matplotlib.

**Pandas:** usada para analisar, manipular e visualizar dados de forma eficiente. Ele é amplamente utilizado em ciência de dados, análise estatística e machine learning.

**Matplotlib:** usada para criar gráficos e visualizações de dados.

```python
# import é importar, pandas é nome da biblioteca e o pd é o "aplelido",
#será com pd que nos vamos chamar a bibilioteca em nosso codigo.

import pandas as pd

# Matplotlib é o nome da biblioteca e o plt é o apelido.
import matplotlib.pyplot as plt
```

```python
"""
Df é DataFrame, a abreviação Df é a mais utilizada então é muito comum você encontrar assim.
pd.read_csv() - Imagina que é uma fraze o pd é qual biblioteca nos estamos chamando, como ja vimo pd é o "apelido" do pandas,
read é a ação que o pd vai fazer, ou seja ele vai ler alguma coisa,
csv é o formato do arquivo que ele vai ler,
dentro do () entra o objeto que ele vai ler, no nosso caso é "Alunos.csv" que a partir de agora será nosso Df.

"""
Df = pd.read_csv('Alunos.csv', sep= ";")
```

```python
"""
Já sabemos que o Df é nossa DataFrame,
o .shape é a ação, ou seja, ele vai retornar o tamanho do nosso Df,
no caso, 100 linhas e 6 colunas.
"""

Df.shape
```

```python
"""
A função .info() é uma das mais importantes, porque ela retorna quais variaves nos temos,
o Dtype (Tipo da variavel) e se a coluna tem valores nulos ou não.

(Importante) Valores nulos são uma dificuldade dentro da analise de dados, pode prejudicar a vizualização e a construção de novas metricas,
 existe formas de tratar esses valores, mas não vamos ver isso hoje.
"""
Df.info()
```

```python
"""
A função describe() ajuda a entender a dispoção dos graficos de forma mais quantitativa,
ajuda a indicar o outliners e a distribuição dos dados.

(Exmplo) Se na coluna idade o valor maximo fosse 200 já identificariamos uma informação suspeita no Df.
"""
Df.describe()
```

Finalizando a analise exploratoria, vamos iniciar a faze de pré-porcessamento:
* Muitas vezes nessa etapa precisamos adequar o tipo da variavel
* Tratar os valores nulos
* Eliminar os outliners

Mas nessa caso estamos com a estrutura da tabela já pronta então,
vamos perguntar, o que eu posso agregar, quais informações eu consigo cruzar gerando uma nova medida, isso é a engenharia de analytics.

```python
'''
Vamos criar a media das notas de portugues e matematica,

primeiro passo nomear a variavel ( coloque nomes que você vai se lembrar depois ),

Media_Matematica será a nossa, ela vai receber qual valor?
vamos buscar no nosso Df que se chama Df, na coluna "Nota de Matematica" por isso Df["Nota de Matematica],
Qual ação vamos usar?
.mean() que retorna a média dos valores.

Vamos fazer quasse a mesma coisa para a coluna de portugues.
nome da variavel: Media_Portugues.
O que ela recebe? Df["Nota de Portugues"].
Qual ação? .mean().

'''
Media_Matematica = Df['Nota de Matematica'].mean()
Media_Portugues = Df['Nota de Portugues'].mean()
```

```python
"""
Agora vamos criar uma variavel que se origina a partir do cruzamento de outras duas,
primeiro vamos separa em dois grupos, o Df_fem que vai receber todas as linhas que na coluna sexo seja "Female",
e o Df_masc que vai receber todas as linhas que na coluna sexo seja "Male".
"""

Df_fem = Df[Df['Sexo'] == "female"]
Df_masc = Df[Df['Sexo'] == "male"]

"""
Agora a partir dessa variavel vamos fazer a media das notas de Matematica e portugues,
Variavl: Media_Matematica_F
o que ela recebe? Df_fem["Nota de Matematica"]
Qual ação? .mean()

"""
Media_Matematica_F = Df_fem['Nota de Matematica'].mean()
Media_Matematica_M = Df_masc['Nota de Matematica'].mean()

Media_Portugues_F = Df_fem['Nota de Portugues'].mean()
Media_Portugues_M = Df_masc['Nota de Portugues'].mean()
```

Agora com as variaveis prontas, podemos ir para o proximo passo, a vizualização dos dados.

Nessa etapa o mais importante é gerar imagem que facilite o entendimento dos dados e possibilite a criação de concluções uteis para seu problema.

Lembrando que a "não relação" tambem pode ser uma informação util. (vamos tratar disso na primeira imagem)

```python
"""
Em primeiro momento a criação de grafico a partir do texto nos é estranha por não ser comum.

A principal vantagem de ser assim é a liberdade que o analista ganha ao poder configurar seu proprio grafico,
algo que é muito mais limitados em programas como Excel ou power BI.

É o portante perceber que cada linha é um elemento do grafico,
comprender linha por linha é essencial para entender o grafico como um todo
e o mais importante, entendo qual elemento é em cada linha você pode adaptar o grafico de acordo com a sua necessidade.
"""

# Dados para o gráfico
medias = [Media_Matematica, Media_Portugues]
disciplinas = ['Matemática', 'Português']

# Criar o gráfico de barras
plt.bar(disciplinas, medias) # Aqui são os valores que sera "plotado" no grafico, e o tipo do grafico.

# Adicionar título e rótulos aos eixos
plt.title('Média das Notas') #Titulo
plt.xlabel('Disciplina') #Rotulo eixo X
plt.ylabel('Média') #Rotulo eixo Y

# Exibir o gráfico
plt.show()
```

```python
"""
Repare que o texto para criar esses dois graficos é masicamente o mesmo, a diferença é em qual variavel estamos trabalhando.
"""
# Dados para o gráfico
medias = [Media_Matematica_F, Media_Matematica_M]
Sexo = ['Feminino' , 'Masculino']

# Criar o gráfico de barras
plt.bar(Sexo, medias)

# Adicionar título e rótulos aos eixos
plt.title('Média das Notas de Matematica')
plt.xlabel('Sexo')
plt.ylabel('Média')

# Exibir o gráfico
plt.show()
```

Como podemos ver nos graficos, não existe grandes diferenças entre os dois grupos, nesse caso, pode comprovar a "não relação" entre sexo e nota.

A "não relação" é uma informação tão importante quanto a relação.

Agora que entendos os principios de montagem do grafico, vamos adicionar um grau de complexidade, nessa etapa vamos unir a criação de uma nova vaiavel com a criação do grafico.

```python
"""

Pensando em codigo temo o "groupby", a função dele é agrupar os dados de acordo com uma coluna especifica
e a ação .cont que faz a contagem daquela variavel.

Adicionei tambem uma ação para modificar a estrutura da tabela, vamos mudar o nome da coluna "Nome" para "Quantidade de Alunos".

"""
# Agrupar por idade e contar alunos
idade_alunos = Df.groupby('Idade')['Nome'].count().reset_index()  # Converte para DataFrame

# Renomear colunas corretamente
idade_alunos.columns = ['Idade', 'Quantidade de Alunos']
```

```python
# Criar o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(idade_alunos['Idade'], idade_alunos['Quantidade de Alunos'])
plt.title('Dispersão entre Idade e Quantidade de Alunos')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Alunos')
plt.grid(True)  # Adiciona grade ao gráfico
plt.show()  # Exibe o gráfico
```

```python
"""
Agora vamos fazer quase a mesma coisa, vamos apenas adicionar mais um recorte, quero saber a quantidade de alunos por sexo.

Então vamos agrupar por idade e sexo, contando alunos e vamos adicionar 3 linhas no codigo do grafico.
"""
# Agrupar por idade e sexo, contando alunos
idade_sexo_alunos = Df.groupby(['Idade', 'Sexo'])['Nome'].count().reset_index()
idade_sexo_alunos.rename(columns={'Nome': 'Quantidade de Alunos'}, inplace=True)

# Separar dados de masculinos e femininos
masculinos = idade_sexo_alunos[idade_sexo_alunos['Sexo'] == 'male']
femininos = idade_sexo_alunos[idade_sexo_alunos['Sexo'] == 'female']

# Criar o gráfico de dispersão com cores diferentes
plt.figure(figsize=(8, 6))
plt.scatter(masculinos['Idade'], masculinos['Quantidade de Alunos'], color='blue', label='Masculinos') # Quero que os pontos que se referem ao masculino sejam azul
plt.scatter(femininos['Idade'], femininos['Quantidade de Alunos'], color='red', label='Femininos') # Quero que os pontos que se referem ao feminino sejam vermelho
plt.title('Dispersão entre Idade e Quantidade de Alunos por Sexo')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Alunos')
plt.grid(True)
plt.legend()  # Adiciona legenda para identificar as cores
plt.show()
```
